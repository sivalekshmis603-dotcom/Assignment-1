```python
num1 = int(input("Enter the first integer: "))
num2 = int(input("Enter the second integer: "))


addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (modulus by zero)"
exponent = num1 ** num2

print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
print(f"Modulus: {num1} % {num2} = {modulus}")
print(f"Exponentiation: {num1} ** {num2} = {exponent}")
```

```python
import numpy as np
numbers = [12, 45, 23, 89, 4, 67, 33, 21, 90, 56]

print(f"Length of the list: {len(numbers)}")
      
print(f"Maximum value: {max(numbers)}")
print(f"Minimum value: {min(numbers)}")

numbers.append(100)
print(f"List after adding 100: {numbers}")

numbers.remove(23)
print(f"List after removing 23: {numbers}")


ascending = sorted(numbers)
print(f"Sorted in ascending order: {ascending}")

descending = sorted(numbers, reverse=True)
print(f"Sorted in descending order: {descending}")

arr = np.array(numbers)

mean_val = np.mean(arr)
median_val = np.median(arr)
std_val = np.std(arr)

print("\n--- NumPy Calculations ---")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Standard Deviation: {std_val}")
```

```python

student = {
    "name": "Rahul",
    "age": 21,
    "course": "Data Science",
    "marks": 88
}

print("Student Details:")
for key, value in student.items():
    print(f"{key}: {value}")

student["grade"] = "A"
print("\nUpdated Dictionary with Grade:")
print(student)

courses = {"Python", "Data Science", "AI", "Python"}
print("\nUnique Courses Set:")
print(courses)

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print("\nUnion of sets:", set1 | set2)
print("Intersection of sets:", set1 & set2)
print("Difference of sets (set1 - set2):", set1 - set2)
print("Difference of sets (set2 - set1):", set2 - set1)
```

```python

data = """Rahul, Data Science, 82
Anjali, Python, 67
Mohit, AI, 91
Sneha, Machine Learning, 74
Arjun, Cyber Security, 88"""

with open("student_data.txt", "w") as file:
    file.write(data)

print("Data written to student_data.txt successfully!")

print("\nStudents with marks above 75:")

with open("student_data.txt", "r") as file:
    for line in file:
        name, course, marks = line.strip().split(", ")
        marks = int(marks)

        if marks > 75:
            print(f"Name: {name}, Course: {course}, Marks: {marks}")
```
